```python
import warnings
warnings.filterwarnings('ignore')
from jupyter_dash import JupyterDash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output 
import pandas as pd
import plotly.express as px 
```

```python
df = pd.read_csv('avocado-updated-2020.csv') #read csv
df.info()
```

```python
fig = df['geography'] == 'Boston'
px.line(df[fig] , x= 'date' , y='average_price',color ='type')
```

```python
app = JupyterDash() #create new dashboard called app

#Layout the app 
# first div => inside the div:
# 1- H1 title 
# 2 Pargraph 
# 3- Dropdown menu whit data from dataframe (geography)
# 4- Graph to hold the plotly figure

app.layout = html.Div(id='main_div',

    # Childern of div 
    children=[
        html.H1('Avocado Price',style={'textAlign': 'center', 'fontSize':50, 'color':'#25450C','fontFamily':'Trebuchet MS'}),
        html.P('This is a dash bord to show a prices of avocado in different state from 2015 to 2020',style={'textAlign': 'center', 'color':'#25450C','fontFamily':'Trebuchet MS'}),
        dcc.Dropdown(id='country_list',
                     options=[{'label':i, 'value':i }for i in df['geography'].unique()], style={'width': '80%', 'fontFamily':'Trebuchet MS','margin':'auto','color':'#25450C'}),      
        dcc.Graph(id='graph')
    ],

    # Styel for the div element 
    style={'background-color': '#FBF8D9',
          'background-size': '100%',
          'position': 'fixed',
          'width': '100%',
          'height': '100%'
          }
)

# add callback to update the graph after each click
@app.callback(
        
    Output('graph','figure'),
    Input('country_list','value')
)

# update function to display the data based on the selected geography 
def update_graph(selected_country):

    filtered_df = df[df['geography']==selected_country]
    line_chart = px.line(filtered_df, x='date',y='average_price', color='type',template='ggplot2',height=450 , width=1500)
    line_chart.update_layout(margin=dict(l=50, r=50, t=50, b=50),paper_bgcolor="#FBF8D9")
    return line_chart

if __name__ == '__main__':
    app.run_server()
```
